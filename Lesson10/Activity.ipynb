```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

housing_df = pd.read_csv('HousingData.csv')
sns.set()
%matplotlib inline
```

```python
x = housing_df['ZN']
y = housing_df['MEDV']
plt.scatter(x, y)
plt.show()
```

```python
plt.figure(figsize=(10, 7))
sns.regplot(x,y)
plt.show()
```

```python
housing_df['ZN'] = housing_df['ZN'].fillna(housing_df['ZN'].mean())
housing_df['MEDV'] = housing_df['MEDV'].fillna(housing_df['MEDV'].mean())
```

```python
import statsmodels.api as sm
X = sm.add_constant(x)
model = sm.OLS(y, X)
est = model.fit()
print(est.summary())
```
