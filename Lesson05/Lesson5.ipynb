```python
# Start the definition of the funtions with the keyword 'class'
# Immediately after keyword, the name of the class is included with colon.
class Person:

    # The __init__() function, as mentioned is essential in user defined functions
    # Regardless if the classes contains this function, in the backend it will be used
    # In this case, we will use this function for initializing the following for an object:
    # 1. Name
    # 2. Age
    # 3. Gender
    # 4. Hobby
    # Self parameter is referring the class itself
    def __init__(self, name, age, gender, hobby):
        
        # Name of the object
        self.name = name
        
        # Age of the object
        self.age = age
        
        # Gender of the object
        self.gender = gender
        
        # Hobby of the object
        self.hobby = hobby

# Here p1 and p2 are the objects of the class Person
# p1 and p2 are initialized with features for the class Person
p1 = Person("Joe", 31, 'male', 'fishing')
p2 = Person("Mary", 24, 'female', 'skating')
```

```python
import math

# Start the definition of the funtions with the keyword 'class'
# Immediately after keyword, the name of the class is included with colon.
class Circle:

    # The __init__() function, as mentioned is essential in user defined functions
    # Regardless if the classes contains this function, in the backend it will be used
    # In this case, we will use this function for initializing the following for an object:
    # 1. Radius
    # Self parameter is referring the class itself
    def __init__(self, radius):
        
        # Radius of the circle getting defined
        self.radius = radius
        
    # Class method to calculate area of the circle
    def calculateArea(self):
        return math.pi*float(self.radius)


# Here c1 is a circle object which is initialized with radius
p1 = Circle(10)

print("Radius of the circle defined is : {} cms".format(p1.radius))
print("Area of the circle defined is : {} cms^2".format(p1.calculateArea()))
```

```python
# Start defining the class and name it Date
class Date:
    
    # This is a static variable which will be used in the class
    # It will come in handy when calculating the days remaining for the weekend
    daysInAWeek = {
        'monday':1,
        'tuesday':2,
        'wednesday':3,
        'thursday':4,
        'friday':5,
        'saturday':6,
        'sunday':7
        
    }
    
    # Initalizing the object with the following variables
    # 1. Day of the month (int)
    # 2. Month (int)
    # 3. Year (int)
    # 4. Day of the week (string)
    def __init__(self, dayOfMonth, month, year, dayOfTheWeek):
        self.dayOfMonth = dayOfMonth
        self.month = month
        self.year = year
        self.dayOfTheWeek = dayOfTheWeek
        
    # First required function to print the date in correct format
    def printFormatedDate(self):
        print('{}/{}/{} {}'.format(self.month, self.dayOfMonth, self.year, self.dayOfTheWeek))
    
    # Second required function to get the number of days left for Sunday
    def getNumberOfDaysTillSunday(self):
        
        # As mentioned above, using the static dictionary
        currentDayOfWeek = self.daysInAWeek[self.dayOfTheWeek.lower()]
        daysLeft = 7-currentDayOfWeek
        
        # Days left are more that 0, then print the number of days left
        if daysLeft > 0:
            print('Number of days left for Sunday are: {}'.format(daysLeft))
        # In case the number of days are 0, then today is Sunday
        else:
            print('Today is Sunday!')
        
# Initalize Date object with day of the month, month, year and day of the week
d1 = Date(9,3,2019, 'Saturday')

# Function #1, to print the date in given format
d1.printFormatedDate()

# Function #2, to print the number of days for Sunday
d1.getNumberOfDaysTillSunday()

# Similar example, but give date is a Sunday
d2 = Date(9,11,2019, 'Sunday')
d2.printFormatedDate()
d2.getNumberOfDaysTillSunday()
```

```python
# Library for defining date
import datetime

# Class definition for car
class Car:
    
    # Constructor for car with the required features
    # Here we can have many more features to define the car even better
    # Example: Inital miles (so show it was a used car),
    # isRental to show if it was a rental car etc.
    def __init__(self, make, model, year, bought, 
                 buyingPrice, mileage, sellingPrice, sellingDate):
        self.make = make
        self.model = model
        self.year = year
        self.bought = bought
        self.buyingPrice = buyingPrice
        self.mileage = mileage
        self.sellingPrice = sellingPrice
        self.sellingDate = sellingDate
    
    # Print the information of the car in a good format
    # As mentioned in earlier chapters, inbuilt function 'format' is very useful in these cases
    def printInformation(self):
        
        print('Following is the information about the car:')
        print('Make: {}'.format(self.make))
        print('Model: {}'.format(self.model))
        print('Year: {}'.format(self.year))
        print('Bought on: {}'.format(self.bought))
        print('Bought for: ${}'.format(self.buyingPrice))
        print('Mileage: {} miles'.format(self.mileage))
        print('Sold for: ${}'.format(self.sellingPrice))
        print('Sold on: {}'.format(self.sellingDate))
        
    # Calculate depreciation
    # For simplicity, we will be concentrating only on the amount depreciated per mile
    # We will also be calculating the time needed to depreciate that amount
    def depreciationPerMile(self):
        
        depreciation = self.buyingPrice - self.sellingPrice
        depreciationRate = depreciation/self.mileage
        
        differenceDate = self.sellingDate - self.bought
        print("\nFollowing is the information regarding the depreciation of the car")
        print('The car has depreciated from ${} to ${} in {} miles at the rate of {} $/mile in {} days'.\
                                                              format(self.buyingPrice, 
                                                                self.sellingPrice, 
                                                                self.mileage, 
                                                                round(depreciationRate, 3), 
                                                                differenceDate))

# Object of car with the required features
c1 = Car('Honda', 'Civic', '2016', 
         datetime.datetime(2016,12,9), 
         24000, 36000, 16500, 
         datetime.datetime(2018,10,25))

# Above defined functions
c1.printInformation()
c1.depreciationPerMile()        
```

```python
# Make class Polygol which is defined as a geometric structure with certain number of sides
class Polygon:
    
    # For initializing polygon, we need number of sides
    def __init__(self, numberOfSides):
        self.numberOfSides = numberOfSides
        self.sides = []
    
    # A method within the class to set the lengths
    def setSides(self, lengthList):
        for i in range(0,self.numberOfSides):
            self.sides.append(float(lengthList[i]))
    
    # Print the polygon properties
    def printSides(self):
        print('Given polynomial has {} sides'.format(self.numberOfSides))
        print('Following are the lengths of each sides:')
        for i in range(0, self.numberOfSides):
            print(self.sides[i])
            
# A derived class which inherits the features from Polygon
class Square(Polygon):
    
    # We initialize an inherited class using __init__() function of base class
    # Here we can put in more features if needed
    def __init__(self):
        Polygon.__init__(self, 4)
    
    # Check if the given features are that of a square
    # Simple test would be checking the length of all sides is equal
    def checkIfSquare(self):
        if len(list(set(self.sides))) == 1:
            return True
        else:
            return False
    
    # We will be running the calculation only if the structure is a valid Square
    # Area of square is side*side or side^2
    # Perimeter is given by 4*side
    # Since all four sides are going to be equal, we can take anyone for calculations
    def findAreaAndPerimeter(self):
        if self.checkIfSquare():
            area = self.sides[0]*self.sides[0]
            perimeter = 4*self.sides[0]
            print('The area of the square is {}'.format(area))
        else:
            print('This is not a square')
            
print('First square test:')
# Initalize square class, it will get properties from polygon
s1 = Square()

# Method of class Polygon, inherited
s1.setSides([4,4,4,4])

# Method of class Polygon, inherited
s1.printSides()

# Method of class Square
s1.findAreaAndPerimeter()

print()
# Test to show the conditional statement of finding whether given properties are of a square
print('Second square test:')
s1 = Square()
s1.setSides([4,1,4,4])
s1.printSides()
s1.findAreaAndPerimeter()
```

```python
# Base class 1
class Person(object): 
    
    # Needs only name to be initialized
    def __init__(self, name): 
        self.name = name 

# Base class 2
class ITProfessional(): 
   
    # Needs department and project
    def __init__(self, department, project):
        self.department = department
        self.project = project

# Class employee will be inherited from two above classes
class Employee(Person, ITProfessional):
    
    # In order to run the __init__() functions from above mentioned classes we need following info:
    # 1. Name
    # 2. Department
    # 3. Project
    # Also here, a new instance variable is given, ID
    def __init__(self, ID, name, department, project):
        # Instance variable of class Employee
        self.ID = ID
        # Base class #1 constructor
        Person.__init__(self, name)
        # Base class #2 constructor
        ITProfessional.__init__(self, department, project)
    
    # Print information from gather data
    def printInformation(self):
        # Name is inherited from base class #1
        print('Name of the Employee: {}'.format(self.name))
        # ID is original to class Employee
        print('Employee\'s ID {}'.format(self.ID))
        # Department and Project are inherited from base class #2
        print('Department of the Employee: {}'.format(self.department))
        print('Employee is working on project: {}'.format(self.project))
        

e1 = Employee(12345, 'Alex Lee','Data Science', 'Recommendation Engine')
e1.printInformation()
```

```python
import datetime

class Company:
    
    def __init__(self, name, numberOfEmployees, business):
        self.name = name
        self.numberOfEmployees = numberOfEmployees
        self.business = business
    
    def printCompanyInfo(self):
        print('Company Information:')
        print('Company Name: {}'.format(self.name))
        print('Number of Employees: {}'.format(self.numberOfEmployees))
        print('Type of Business: {}'.format(self.business))
        
class PublicCompany:
    
    def __init__(self, stockPrice, publicDate, initialIPO):
        self.stockPrice = stockPrice
        self.publicDate = publicDate
        self.initialIPO = initialIPO
        
    def printPublicInfo(self):
        print('Public information of the company')
        print('Today\'s stock price: ${}'.format(self.stockPrice))
        print('The company went public on {}'.format(self.publicDate))
        print('Initial evaluated price of the IPO: ${}'.format(self.initialIPO))
        
class myCompany(Company, PublicCompany):
    
    def __init__(self, nameOfCEO, nameOfCTO, name, 
                 numberOfEmployees, business, stockPrice, 
                 publicDate, initialIPO):

        self.nameOfCEO = nameOfCEO
        self.nameOfCTO = nameOfCTO
        Company.__init__(self, name, numberOfEmployees, business)
        PublicCompany.__init__(self, stockPrice, publicDate, initialIPO)
        
    def printMyCompany(self):
        
        print('Following are the top officials in my Company:')
        print('Name of the Chief Executive Officer: {}'.format(self.nameOfCEO))
        print('Name of the Chief Technology Officer: {}'.format(self.nameOfCTO))

    def getRateOfDifferenceIPO(self):
        differenceInPrice = self.stockPrice - self.initialIPO
        today = datetime.datetime.now()
        differenceInDate = int((today - self.publicDate).days)
        rateOfDifference = round(differenceInPrice/differenceInDate, 3)
        
        print('The change in IPO price: ${} in {} days at the rate of {} $/days'\
              .format(differenceInPrice,differenceInDate, rateOfDifference))
        
m1 = myCompany('Alex Lee', 'Max Smith', 'Showcase Solutions LLC', 
               1000, 'Motion Pictures Technology', 25, 
               datetime.datetime(2016, 9, 3), 15)

m1.printCompanyInfo()
m1.printPublicInfo()
m1.printMyCompany()
m1.getRateOfDifferenceIPO()
```
